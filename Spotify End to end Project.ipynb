### Converting the feature to vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000)
```

```python
vector=cv.fit_transform(Spotify_df_Modify['feature']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

### Checking Simlarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector)
```

```python
cos_sim=cosine_similarity(vector)
```

```python
cos_sim.shape
```

```python
cos_sim[0]
```

```python
Spotify_df_Modify['song_name']
```

```python
def recommend (song):
    song_index= Spotify_df_Modify[Spotify_df_Modify['song_name']==song].index[0]
    distances= cos_sim[song_index]
    song_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:21]

    for i in song_list:
        print(Spotify_df_Modify.iloc[i[0]].song_name)
```

```python
recommend('Goodbye')
```
